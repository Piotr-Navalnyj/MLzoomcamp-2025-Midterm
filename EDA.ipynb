```python
df = pd.get_dummies(df, columns=['main_position'], drop_first=True)
```

```python
df = df.astype({col: int for col in df.select_dtypes(include=['bool']).columns})
```

```python
n = len(df)
np.random.seed(42)
shuffled_idx = np.random.permutation(n)

```

```python
n_train = int(0.6 * n)
n_val = int(0.2 * n)
n_test = n - n_train - n_val
```

```python
train_idx = shuffled_idx[:n_train]
val_idx   = shuffled_idx[n_train:n_train+n_val]
test_idx  = shuffled_idx[n_train+n_val:]
```

```python
df_train = df.iloc[train_idx].reset_index(drop=True)
df_val   = df.iloc[val_idx].reset_index(drop=True)
df_test  = df.iloc[test_idx].reset_index(drop=True)

```

```python
y_train = np.log1p(df_train["value_eur"].values)
y_val   = np.log1p(df_val["value_eur"].values)
y_test  = np.log1p(df_test["value_eur"].values)
```

```python
X_train_raw = df_train.drop("value_eur", axis=1).values
X_val_raw   = df_val.drop("value_eur", axis=1).values
X_test_raw  = df_test.drop("value_eur", axis=1).values

ones_train = np.ones((X_train_raw.shape[0], 1))
ones_val   = np.ones((X_val_raw.shape[0], 1))
ones_test  = np.ones((X_test_raw.shape[0], 1))

X_train = np.hstack((ones_train, X_train_raw))
X_val   = np.hstack((ones_val, X_val_raw))
X_test  = np.hstack((ones_test, X_test_raw))

print(X_train.shape)
print(X_val.shape)
print(X_test.shape)
```

```python
theta = np.linalg.pinv(X_train) @ y_train


theta.shape
```

```python
y_pred_val_log = X_val @ theta
y_pred_test_log = X_test @ theta
```

```python
y_pred_val = np.expm1(y_pred_val_log)
y_pred_test = np.expm1(y_pred_test_log)
```

```python
y_pred_val = np.clip(y_pred_val, 0, None)
y_pred_test = np.clip(y_pred_test, 0, None)
```

```python
def rmse(y, y_pred):
    return np.sqrt(((y - y_pred) ** 2).mean())

def mae(y, y_pred):
    return np.mean(np.abs(y - y_pred))

def r2(y, y_pred):
    ss_res = ((y - y_pred)**2).sum()
    ss_tot = ((y - y.mean())**2).sum()
    return 1 - ss_res/ss_tot

y_true_val = df_val["value_eur"].values
y_true_test = df_test["value_eur"].values

print("Validation RMSE:", rmse(y_true_val, y_pred_val))
print("Validation MAE:", mae(y_true_val, y_pred_val))
print("Validation R²:", r2(y_true_val, y_pred_val))

print("Test RMSE:", rmse(y_true_test, y_pred_test))
print("Test MAE:", mae(y_true_test, y_pred_test))
print("Test R²:", r2(y_true_test, y_pred_test))
```

```python
y_train_xgb = df_train["value_eur"].values
y_val_xgb   = df_val["value_eur"].values
y_test_xgb  = df_test["value_eur"].values

X_train_xgb = df_train.drop("value_eur", axis=1).values
X_val_xgb   = df_val.drop("value_eur", axis=1).values
X_test_xgb  = df_test.drop("value_eur", axis=1).values
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt


def evaluate(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    return mae, rmse, r2


results = pd.DataFrame({
    "Model": ["Linear Regression", "Random Forest", "XGBoost"],

    "MAE": [
        evaluate(y_test, y_pred_test)[0],       # your LR predictions
        evaluate(y_test, y_pred_test_rf)[0],    # RF
        evaluate(y_test, y_pred_test_xgb)[0],   # XGB
    ],

    "RMSE": [
        evaluate(y_test, y_pred_test)[1],
        evaluate(y_test, y_pred_test_rf)[1],
        evaluate(y_test, y_pred_test_xgb)[1],
    ],

    "R²": [
        evaluate(y_test, y_pred_test)[2],
        evaluate(y_test, y_pred_test_rf)[2],
        evaluate(y_test, y_pred_test_xgb)[2],
    ]
})

print("\n===== FINAL MODEL COMPARISON =====\n")
print(results)


```
